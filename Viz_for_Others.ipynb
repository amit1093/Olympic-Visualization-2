## Olympics Dataset

The dataset is about Olympics data from Year 1896 to 2012. The dataset was formed by combining two datasets :

Dictionary.csv : This has the Country Name, Country Code, GDP per capita and the Population

Summer.csv : This has all the details about the olympics including country code, hosting city, athlete name, sporting events and the medals won.

ISO_Codes.csv: This has the ISO codes for all the countries in the world 

The Dataset formed by combining these three datasets is called OLYMPICS Dataset. It gives details about the hosting city, Countries participating in Olympics, the sports that were there in the Olympics for every Olympic year, the disciplines in every sport, Name of Athletes participating in the events, name of the vent for every discipline of every sports and the medals won by each athlete, the population of Country participating, the per Capita GDP for that country and the respective ISO code.

More than 35,000 medals have been awarded at the Olympics since 1896. The first two Olympiads awarded silver medals and an olive wreath for the winner, and the IOC retrospectively awarded gold, silver, and bronze to athletes based on their rankings. This dataset includes a row for every Olympic athlete that has won a medal since the first games.

Data was provided by the IOC Research and Reference Service and published by The Guardian's Datablog.The dataset was obtained from Kaggle.com. It is an open dataset with authority to edit, update and modify the dataset. There is no license restriction on the dataset and can be used for recreation, editing, downloaing and distribution.

The url for the base datasets is :

- https://www.kaggle.com/the-guardian/olympic-games#summer.csv
- https://www.kaggle.com/the-guardian/olympic-games#dictionary.csv

### GDP per Capita and Medals won for top 20 countries

```python
#finding the GDP of top 20 countries
gdp_20 = pd.DataFrame(olympics.groupby(['Country','GDP per Capita','Population',
                                        'Medal'])['Medal'].count().unstack(-1).sort_values(by = ['Gold'],
                                        ascending = False).head(20))
```

```python
gdp_20 = gdp_20.reset_index(drop = False)
gdp_20
```

```python
Gold = go.Scatter(
                    x = gdp_20['Country'],
                    y = gdp_20['GDP per Capita'],
                    mode='markers',
                    marker=dict(
                        size= 0.05*gdp_20['Gold'],
                        color='Gold',
                        ), name = 'Gold',
                    text = gdp_20['Gold']
)

Silver = go.Scatter(
                    x = gdp_20['Country'],
                    y = gdp_20['GDP per Capita'],
                    mode='markers',
                    marker=dict(size= 0.04*gdp_20['Silver'],
                        color='Grey'),
                    name = 'Silver',
                    text = gdp_20['Silver']
)

Bronze = go.Scatter(
                    x = gdp_20['Country'],
                    y = gdp_20['GDP per Capita'],
                    mode='markers',
                    marker=dict(size= 0.03*gdp_20['Bronze'],
                        color='Burlywood'),
                    name = 'Bronze',
                    text = gdp_20['Bronze']
)

layout= go.Layout(
                    title= 'GDP per Capita vs Country by Medals ',
                    xaxis= dict(title= 'Country'),
                    yaxis=dict(
                    title= 'GDP per Capita')
                
)
                                 

data = [Gold,Silver,Bronze]
    
fig_gdp20 = dict(data = data, layout = layout)
py.iplot(fig_gdp20, filename  = 'gdp per capita scatter')
```

In the above visualization we plot the GDP per capita and the country. The size of the points are based on the medal count of the country. We can see there is a clear decreasing trend in the plot. The countries with greater GDP per capita have greater number of medals showing that the development and economic situations in a country have an affect on the olympics performance.

### Medals won by each country over the years

```python
Country_Medals = olympics.groupby(['Country','Year'])['Medal'].count()
Country_Medals = Country_Medals.unstack(level = -1)
```

```python
Country_Medals = Country_Medals.fillna(0)
Country_Medals.index.values
```

```python
# Dropdown for selecting the countries
Widget1 = ipywidgets.Dropdown(
    options = Country_Medals.index.values,
    value = 'United States',
    description = 'Country')

# Bar plot showing the medal count over the years
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()

tooltip = bqplot.Tooltip(fields = ["x", "y"], labels = ['Year','Medal Count'])

bars = bqplot.Bars(x = Country_Medals.columns.values,y = Country_Medals.loc[Widget1.value].values,scales = {'x':x_sc ,'y': y_sc}, 
                   tooltip = tooltip)

ax_x = bqplot.Axis(scale = x_sc, label = 'Olympic Year')
ax_y = bqplot.Axis(scale = y_sc, label = 'Medal Count', orientation = 'vertical')

def medal_year(change):
    bars.y = Country_Medals.loc[Widget1.value].values
    
Widget1.observe(medal_year)

fig1 = bqplot.Figure(marks = [bars], axes = [ax_x, ax_y],title='Medals won by countries over the years')
fig1.layout.width = '1000px'

display(Widget1, fig1)
```

In the above visualization we plot the total number of medals won by each participating country in each olympics for all the olympics occured. It is an interactive visualization where the country for which the medal count needs to be displayed is selected from the dropdown and this visualization was made with bqplot and ipywidgets. The tooltip displays the year and the total number of medals won in that particular year.

### Choropleth World map based on GDP per Capita

```python
# Making a dictionary of ISO codes and GDP per capita to provide as color input for the map 
gdp_country = {}

for i,j in zip(olympic_final['iso'].unique(), olympic_final['GDP per Capita'].unique()):
    gdp_country[i.astype(int)] = j
```

```python
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()

lines = bqplot.Lines(x = Country_Medals.columns.values,y = [],scales = {'x':x_sc ,'y': y_sc}, colors = ['Red'])

ax_x = bqplot.Axis(scale = x_sc, label = 'Olympic Year')
ax_y = bqplot.Axis(scale = y_sc, label = 'Medal Count', orientation = 'vertical')

fig = bqplot.Figure(marks = [lines], axes = [ax_x, ax_y])
fig.layout.width = '1000px'
```

```python
c_sc = bqplot.ColorScale(scheme = 'RdYlBu')

map_styles = {'color' : gdp_country,
                'scales': {'projection': bqplot.Mercator(), 'color': c_sc}
              } 

world_map = bqplot.Map(map_data=bqplot.topo_load('map_data/WorldMap.json'),
                       ref_data = olympic_final,
                       **map_styles,tooltip_widget = fig,
                       colors={'default_color':'#eef3fe'},
                       hovered_styles= {'hovered_fill': 'White', 'hovered_stroke': 'Black', 'hovered_stroke_width':4.0})


axis = ColorAxis(scale=c_sc,tick_style={'font-size': '12px','font-weight':'bold','font-family':'serif'})

map_fig = bqplot.Figure(marks=[world_map],title_style={'font-size': '20px','font-weight':'bold','font-family':'serif'},
                        title='World Map Based on GDP per Capita',
                        axes=[axis],background_style={'fill':'lightblue'})


map_fig.fig_margin={'top':0,'bottom':0,'left':0,'right':0}

#Defining interations for the map
world_map.interactions = {'hover': 'tooltip'}
```

```python
# Defining the hover handler to access data from the map by hovering
hovered_symbol = ''
def hover_handler(self, content):
    global hovered_symbol
    symbol = content.get('data', '')
    
    if(symbol['name'] != hovered_symbol):
        hovered_symbol = symbol['name']
        if hovered_symbol in  Country_Medals.index.values:
            if(Country_Medals.loc[hovered_symbol] is not None):
                lines.y = Country_Medals.loc[hovered_symbol].values
                fig.title = ('Medal Count over the Years for '+ hovered_symbol)

world_map.on_hover(hover_handler)


b2 = ipywidgets.VBox
display(b2([map_fig,fig]))
```

The above visualization is a Choropleth world map based on the GDP per capita income. This is an interactive visualization where the total medal won by a country over the olympic years is displayed as line plot below the world map by hovering on the country. The countries which didn't won a medal are displayed in white and the reamining countries are color coded based on their GDP percapita income. RdYlBu - Red, Yellow and Blue is the color schema used and the countries with less GDP are in red color and the countries with higher GDP are in Yellow and Blue. The shade of the color indicates the GDP range in the same color bracket.
